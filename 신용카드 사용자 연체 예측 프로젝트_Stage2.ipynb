ÏïåÎ∞îÌé´ ÏàúÏÑúÏóê Îî∞Îùº Î¨∏ÏûêÌòï Îç∞Ïù¥ÌÑ∞Î•º uniqueÌïú ÏàòÏπòÎ°ú Îß§ÌïëÌïòÎäî Í∏∞Î≤ïÏûÖÎãàÎã§.

Î¨∏ÏûêÎ•º Ï†ïÏàòÎ°úÎßå Î≥ÄÌôòÌïòÍ∏∞ ÎïåÎ¨∏Ïóê Îç∞Ïù¥ÌÑ∞ ÌÅ¨Í∏∞ÏôÄ shapeÏù¥ Î≥ÄÌïòÏßÄ ÏïäÏäµÎãàÎã§.

```python
import pandas as pd
import numpy as np

data = {
    "Temperature": ["Hot", "Cold", "Very Hot", "Warm", "Hot", "Warm", "Warm", "Hot", "Hot", "Cold"],
    "Color": ["Red", "Yellow", "Blue", "Blue", "Red", "Yellow", "Red", "Yellow", "Yellow", "Yellow"],
    "Target": [1, 1, 1, 0, 1, 0, 1, 0, 1, 1]}

df = pd.DataFrame(data, columns = ["Temperature", "Color", "Target"])
df
```

```python
# Label Encoding
from sklearn.preprocessing import LabelEncoder

print(df['Temperature'].unique())
print()

# LabelEncoderÎ•º Í∞ùÏ≤¥Î°ú ÏÉùÏÑ±
encoder = LabelEncoder()

# fit, transform Î©îÏÜåÎìúÎ•º ÌÜµÌïú Î†àÏù¥Î∏î Ïù∏ÏΩîÎî©
encoder.fit(df['Temperature'])

df["Temp_Label_Encoder"] = encoder.transform(df['Temperature'])
df
```

###### One-Hot Encoding

Ïπ¥ÌÖåÍ≥†Î¶¨Ïùò ÏàòÎßåÌÅº Î∞±ÌÑ∞Í∞Ä ÏÉùÏÑ±Îê©ÎãàÎã§.

Í∑∏Î†áÍ∏∞ ÎïåÎ¨∏Ïóê Ïπ¥ÌÖåÍ≥†Î¶¨Í∞Ä ÎÑàÎ¨¥ ÎßéÏùÄ Î≥ÄÏàòÏùò Í≤ΩÏö∞ cardinalityÍ∞Ä Ï¶ùÍ∞ÄÌï¥ Î™®Îç∏Ïùò ÏÑ±Îä•ÏùÑ Ï†ÄÌïòÏãúÌÇ¨ Ïàò ÏûàÏúºÎãà, Ïú†ÏùòÌï¥Ï£ºÏÑ∏Ïöî.

```python
from sklearn.preprocessing import OneHotEncoder

oh = OneHotEncoder()
encoder = oh.fit_transform(df['Temperature'].values.reshape(-1,1)).toarray() # Ïù∏ÏΩîÎî© ÌïòÍ∏∞ Ï†ÑÏóê 2Ï∞®Ïõê Îç∞Ïù¥ÌÑ∞Î°ú Î≥ÄÌôò

df_OneHot = pd.DataFrame(encoder, columns=["Temp_OneHot_Encoder_" + str(oh.categories_[0][i]) for i in range (len(oh.categories_[0]))])

df_onehot_result = pd.concat([df, df_OneHot], axis=1)
df_onehot_result
```

```python
df_onehot_result = pd.get_dummies(df_onehot_result, prefix=["Temp_Get_Dummies"], columns=["Temperature"])
df_onehot_result
```

```python
replace_df = pd.DataFrame({
    'PhoneNumber': ['Í≥µÏùºÍ≥µ-Í≥µÍ≥µÍ≥µÍ≥µ-Í≥µÍ≥µÍ≥µÍ≥µ', '01000000000']
})

def new_number(number):
    myDict = {
    'Í≥µ': '0',
    'Ïùº': '1',
    'Ïù¥': '2',
    ',' : '',
    '-': '',
    '.': '',
    }
    for k, v in myDict.items():
        number = number.replace(k, v)
    return number

replace_df["processed_number"] = replace_df["PhoneNumber"].apply(new_number)
replace_df
```

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import OneHotEncoder

# Ïù∏ÏΩîÎçî ÏÉùÏÑ±
label_encoder = LabelEncoder()
onehot_encoder = OneHotEncoder()

# Ïù∏ÏΩîÎî©Ìï† Ïó¥ ÏÑ†ÌÉù
columns_to_label = ['income_type', 'edu_type', 'family_type', 'house_type', 'occyp_type']
columns_to_onehot = ['gender', 'car', 'reality']

# ÏÑ†ÌÉùÌïú Ïó¥Ïóê ÎåÄÌï¥ ÎùºÎ≤® Ïù∏ÏΩîÎî© ÏàòÌñâ
for column in columns_to_label:
    # train Îç∞Ïù¥ÌÑ∞Ïóê ÎåÄÌï¥ÏÑúÎäî fit_transform ÏàòÌñâ
    train[column] = label_encoder.fit_transform(train[column])
    print(label_encoder.classes_)

    # test Îç∞Ïù¥ÌÑ∞Ïóê ÎåÄÌï¥ÏÑúÎäî transformÎßå ÏàòÌñâ
    test[column] = label_encoder.transform(test[column])

# ÏÑ†ÌÉùÌïú Ïó¥Ïóê ÎåÄÌï¥ ÏõêÌï´ Ïù∏ÏΩîÎî© ÏàòÌñâ
for column in columns_to_onehot:
    # train Îç∞Ïù¥ÌÑ∞Ïóê ÎåÄÌï¥ÏÑúÎäî fit_transform ÏàòÌñâ
    train[column] = onehot_encoder.fit_transform(train[column].values.reshape(-1, 1)).toarray()
    print(onehot_encoder.categories_)

    # test Îç∞Ïù¥ÌÑ∞Ïóê ÎåÄÌï¥ÏÑúÎäî transformÎßå ÏàòÌñâ
    test[column] = onehot_encoder.transform(test[column].values.reshape(-1, 1)).toarray()
```

Log LossÏùò ÏãùÏùÄ Îã§ÏùåÍ≥º Í∞ôÏäµÎãàÎã§:
$${\displaystyle \operatorname {LogLoss} =-\frac{1}{N}\sum _{i=1}^{N}\sum _{j=1}^{M}y_{ij}\log(p_{ij})}$$
Ïó¨Í∏∞ÏÑú \(N\)ÏùÄ Îç∞Ïù¥ÌÑ∞Ïùò Í∞úÏàò, \(M\)ÏùÄ ÌÅ¥ÎûòÏä§Ïùò Í∞úÏàò, \(y_{ij}\)Îäî iÎ≤àÏß∏ Îç∞Ïù¥ÌÑ∞Ïùò jÎ≤àÏß∏ ÌÅ¥ÎûòÏä§Ïóê ÎåÄÌïú Ïã§Ï†ú Î†àÏù¥Î∏î(1 ÎòêÎäî 0), \(p_{ij}\)Îäî iÎ≤àÏß∏ Îç∞Ïù¥ÌÑ∞Í∞Ä jÎ≤àÏß∏ ÌÅ¥ÎûòÏä§Ïóê ÏÜçÌï† ÏòàÏ∏° ÌôïÎ•†ÏûÖÎãàÎã§.

Log LossÎ•º Í≥ÑÏÇ∞ÌïòÍ∏∞ ÏúÑÌïú log_loss ÎùºÏù¥Î∏åÎü¨Î¶¨Î•º Î®ºÏ†Ä ÏûÑÌè¨Ìä∏Ìï¥Ïïº Ìï¥Ïöî!
