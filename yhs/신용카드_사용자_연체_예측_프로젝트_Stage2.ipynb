```python
from sklearn.preprocessing import LabelEncoder

# ÎùºÎ≤® Ïù∏ÏΩîÎçî ÏÉùÏÑ±
label_encoder = LabelEncoder()

# ÎùºÎ≤® Ïù∏ÏΩîÎî©Ìï† Ïó¥ ÏÑ†ÌÉù
columns_to_encode = ['gender', 'car', 'reality', 'income_type', 'edu_type', 'family_type', 'house_type', 'occyp_type']

# ÏÑ†ÌÉùÌïú Ïó¥Ïóê ÎåÄÌï¥ ÎùºÎ≤® Ïù∏ÏΩîÎî© ÏàòÌñâ
for column in columns_to_encode:
    # train Îç∞Ïù¥ÌÑ∞Ïóê ÎåÄÌï¥ÏÑúÎäî fit_transform ÏàòÌñâ
    train[column] = label_encoder.fit_transform(train[column])
    print(label_encoder.classes_)

    # test Îç∞Ïù¥ÌÑ∞Ïóê ÎåÄÌï¥ÏÑúÎäî transformÎßå ÏàòÌñâ
    test[column] = label_encoder.transform(test[column])
```

ÌèâÍ∑† Ï†úÍ≥±Í∑º Ïò§Ï∞®)Í∞Ä ÌèâÍ∞Ä Í∏∞Ï§ÄÏúºÎ°ú ÏÇ¨Ïö©ÌïòÍ≥† ÏûàÏñ¥Ïöî.

RMSEÎ•º Í≥ÑÏÇ∞ÌïòÍ∏∞ ÏúÑÌïú mean_squared_error ÎùºÏù¥Î∏åÎü¨Î¶¨Î•º Î®ºÏ†Ä ÏûÑÌè¨Ìä∏Ìï¥Ïïº Ìï¥Ïöî!
